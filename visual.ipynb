```python
import codecs
import json
import numpy as np
```

```python
# 对于Reformer，我们要观察是不是相近位置都落入了一个桶中。我们最后要生成一个文本序列，
# 并用渐变色标注其桶编号
with codecs.open('decode_para.txt','r','utf-8') as fin:
    sentence_list = []
    sentence = []
    all_sentence = []
    for line in fin.readlines():
        #print(line)
        for char_ in list(line.strip()):
            sentence.append(char_)
            all_sentence.append(char_)
            if char_ == '。':
                sentence_list.append(sentence)
                sentence = []
    print(sentence_list)
with codecs.open("paragraph.json", 'w', 'utf-8') as f1:
    json.dump(sentence_list, f1, ensure_ascii=False)
```
